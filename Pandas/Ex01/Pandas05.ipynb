##### 1 Matplotlib - 기본 그래프 도구 
- 시각화 도그를 사용하면 데이터의 구조와 패턴을 파악하기가 용이하다.

- 연속하는 데이터의 값들을 직선 또는 곡선으로 연결하여 데이터 값 사이의 관계를 표현

```python
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc
import pandas as pd 

font_path = "C:\\Windows\\Fonts\\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df1 = pd.read_excel("File\\시도별 전출입 인구수.xlsx", engine='openpyxl', header=0)
df1.fillna(method="ffill", inplace=True)
mask = (df1["전출지별"] == "서울특별시") & (df1["전입지별"] !="서울특별시")
df_seoul = df1[mask]
df_seoul =df_seoul.drop(["전출지별"], axis=1)
df_seoul.rename({"전입지별":"전입지"}, inplace=True, axis=1)
df_seoul.set_index("전입지", inplace=True) 

col_year = list(map(str, range(1970, 2018)))
df_3 = df_seoul.loc[["충청남도", "경상북도", "강원도"], col_year]

plt.style.use("ggplot")

fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1, 1, 1)

ax.plot(
    col_year, 
    df_3.loc["충청남도",:], 
    marker='o', 
    markerfacecolor='green', 
    markersize='10', 
    color='olive', 
    linewidth=2, 
    label= '서울= 충남'
    )

ax.plot(
    col_year, 
    df_3.loc["경상북도", :], 
    marker='v', 
    markerfacecolor='red', 
    markersize='10', color='skyblue', 
    linewidth=2, 
    label= '서울= 경북'
    )
ax.plot(
    col_year, 
    df_3.loc["강원도", :], 
    marker='o', 
    markerfacecolor='blue', 
    markersize='10', 
    color='skyblue', 
    linewidth=2, 
    label= '서울= 강원'
    )

ax.legend(loc="best")
ax.set_title("Test", size=20)
ax.set_xticklabels(col_year, rotation=45)
plt.show()
```
