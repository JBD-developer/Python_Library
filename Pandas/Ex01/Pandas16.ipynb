```python
import pandas as pd 
import seaborn as sns 

titanic = sns.load_dataset("titanic")
df = titanic.loc[:, ["age", "sex", "class", "fare", "survived"]]
print(df.head())


pdf1 = pd.pivot_table(df, 
                      index = "class", 
                      columns = "sex", 
                      values = "age", 
                      aggfunc= "mean")

print(pdf1.head(n=10))

# grouped = df.groupby(["class", "sex"])[["age"]].mean()
# print(grouped.head())

pdf2 = pd.pivot_table(df, 
                      index = "class", 
                      columns = "sex", 
                      values = "age", 
                      aggfunc=["mean", "sum"])

print(type(pdf2))
print(pdf2.head(n=10))

pdf3 = pd.pivot_table(df, 
                      index = "class", 
                      columns = "sex", 
                      values = ["age", "fare"],
                      aggfunc = ["mean", "max"])

print(pdf3.index)
print(pdf3.head(n=10))
```
