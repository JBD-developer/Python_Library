```python
import pandas as pd 

ts_ms = pd.date_range(start="2019-01-01",
                      end=None, 
                      periods=6,
                      freq="MS",
                      tz="Asia/Seoul"
                      )
print(ts_ms)

ts_3m = pd.date_range(start="2022-03-29", 
                      end=None,
                      periods=6, 
                      freq="3M",
                      tz="Asia/Seoul"
                      )
print(ts_3m)


pr_y = pd.period_range("2019-01-01",
                       end=None, 
                       periods=3,
                       freq="Y")
print(pr_y)

pr_m = pd.period_range("2019-01-01",
                     end=None, 
                     periods=3, 
                     freq="M")
print(pr_m)

pr_d = pd.period_range("2019-01-01",
                       end=None, 
                       periods=3,
                       freq="D")
print(pr_d)


pr_h = pd.period_range("2019-01-01",
                       end=None,
                       periods=3,
                       freq="H")
print(pr_h)
```

```python
import pandas as pd 

df = pd.read_csv("File\\stock-data.csv")

df["New_Date"] = pd.to_datetime(df.Date)
df["Year"] = df.New_Date.dt.year
df["Month"] = df.New_Date.dt.month
df["Day"] = df.New_Date.dt.day
df["Date_Year"] = df["New_Date"].dt.to_period(freq='Y')
df["Date_Month"] = df["New_Date"].dt.to_period(freq='M')
df["Date_Day"] = df["New_Date"].dt.to_period(freq="D")
print(df.head(n=10))
```
