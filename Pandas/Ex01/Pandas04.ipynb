#### Dataframe의 구조 
##### 01-01. Dataframe 미리보기  
- Dataframe으로 정리된 데이터셋을 처음 접할 때 가장 먼저 head() 메소드이다. Dataframe의 앞부분 일부 내용을 출력한다. 
- 데이터셋의 내용과 구조를 개략적으로 살펴보는 것이 가능하기 때문에 분석 방향을 정하는데 필요한 정보를 얻기 용이하다. 
- Dataframe의 마지막 부분을 보고 싶다면 tail() 메소드를 호출한다. 
- 앞부분 Dataframe.head(n)
- 뒷부분 Dataframe.tail(n)
##### 01-02. 요약 정보 확인하기 
- Dataframe 클래스의 shape 속성은 행과 열의 개수를 튜플 형태로 나타낸다. 
- info() 메소드를 Dataframe에 적용하면 기본 정보를 출력한다.  

```python
import pandas as pd 
pd.set_option('display.max.colwidth', 700)
df = pd.read_csv("File\\auto-mpg.csv", header=None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']

print(df.head(5))
print(df.tail(5))
print(df.shape)
print(df.info())
```
