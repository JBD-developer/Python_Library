```python
import pandas as pd 
import seaborn as sns 

titanic =sns.load_dataset("titanic")
df = titanic.loc[:, ["age", "sex", "class", "fare", "survived"]]

grouped = df.groupby(["class"])
age_mean = grouped.age.mean()
print(age_mean)

age_std = grouped.age.std()
print(age_std)

for key, group in grouped.age: 
    print("origin : " , key)
    group_zscore = (group - age_mean.loc[key]) / age_std.loc[key]
    print(group_zscore.head())
    print("\n")

def z_score(x):
    return (x - x.mean())/x.std()


age_zscore = grouped.age.transform(z_score)
print("\n")
print(age_zscore.head(n=10))

    
```

```python
import pandas as pd 
import seaborn as sns 

titanic = sns.load_dataset("titanic")
df = titanic.loc[:, ["age", "sex", "class", "fare", "survived"]]
grouped = df.groupby(["class"])
grouped_filter1 = grouped.filter(lambda x : len(x) >= 200)
# print(grouped_filter1.head(n=15))
print(len(grouped_filter1))

grouped_filter2 = grouped.filter(lambda x : x.age.mean()< 30)
grouped_filter3 = grouped.apply(lambda x : x.age.mean() < 30)
print(grouped_filter2.head(n=10))
print(len(grouped_filter2))
print(grouped_filter3.head(n=10))
print(len(grouped_filter3))
# print(grouped_filter3)

for x in grouped_filter3.index:
    if grouped_filter3[x] == True : 
        age_filter_df = grouped.get_group(x)
        print(age_filter_df.head())
        print("\n")
```
