```python
import seaborn as sns 

def add_10(n):
    return n + 10 

def add_two_obn(a, b):
    return a + b 

titanic = sns.load_dataset("titanic")
df = titanic.loc[:, ["age", "fare"]]
df["ten"] = 10
print(df.head(10))
sr1 = df["age"].apply(add_10)
print(sr1)

sr2 = df["age"].apply(add_two_obn, b=10)
```

```python
import seaborn as sns 

titanic= sns.load_dataset("titanic")
df = titanic.loc[:, ["age", "fare"]]
print(df.head(10))

def add_10(n):
    return n +10

def missing_value_check(series):
    return series.isnull()

df_map = df.applymap(add_10)
print(df_map.head(n=10)) 
df_check = df.apply(missing_value_check, axis=0)
print(df_check.head(n=10))
```

```python
import seaborn as sns 

titanic = sns.load_dataset("titanic")
df = titanic.loc[:, ["age", "fare"]]
df["ten"] = 10 
print(df.head())

def add_two_obj(a, b):
    return a + b 

df["head"] = df.apply(lambda x : add_two_obj(x["age"], x["ten"]), axis=1)
print(df.head())
```
