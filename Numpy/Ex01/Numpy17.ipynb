```python
import numpy as np 

mat_a = np.arange(9).reshape(3, 3)
print(f"mat_a shape {mat_a.shape} \n{mat_a}")
vec_row, vec_col = mat_a[1, :], mat_a[:, 1]

print(f"vec_row shape {vec_row.shape} {vec_row}")
print(f"vec_col shape {vec_col.shape} {vec_col}")
```
