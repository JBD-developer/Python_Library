```python
# Indexing with int ndarrays 
# Indexing and Slicing Vector ndarrays 
import numpy as np 

vec_a = np.arange(10)
print(f"ndarray:\n{vec_a}\n")

indices = np.array([0, 3, 4])
print(f"vec_a[0, 3, 4] : {vec_a[indices]}")

indices = np.array([0, 0, 5, 5])
print(f"vec_a[0, 0, 5, 5] : {vec_a[indices]}")

indices1 = np.array([[2, 1, 3], [5, 8, 9]])
print(indices1.shape)
print(f"vec_a[[2, 1, 3], [5, 8, 9]] :\n{vec_a[indices1]}")
```

```python
np.random.seed(30)
vec_a = np.random.randint(0, 20, (10, ))
print(f"ndarray:\n{vec_a}\n")

indices = np.random.randint(0, 10, size=(2, 3, 4))
print(indices.shape)
print(indices)
print(vec_a[indices])
# 14, 14, 5, 5
# 13, 13, 13, 12
 
```

```python
mat_a = np.arange(12).reshape(3, 4)
print(f"ndarray: \n{mat_a}\n")
indices = np.array([0, 2])
print(mat_a[indices])

indices = np.array([0, 0, 1, 1, 2, 2])
print(mat_a[indices])

indices = np.array([[0, 1, 2], 
                    [-3, -2, -1]])

print(f"mat_a[[0, 1, 2], [-3, -2, -1]] :\n{mat_a[indices]}")
```

```python
mat_a = np.arange(12).reshape((3, 4))
print(f"ndarray :\{mat_a}")
indices0, indices1 = np.array([0]), np.array([0])
print(f"mat_a[{indices0}, {indices1}] :\n {mat_a[indices0, indices1]}" )

indices0, indices1 = np.array([1]), np.array([1])
print(f"mat_a[{indices0}, {indices1}] :\n {mat_a[indices0, indices1]}" )

indices0, indices1 = np.array([-1]), np.array([0])
print(f"mat_a[{indices0}, {indices1}] :\n {mat_a[indices0, indices1]}" )


indices0, indices1 = np.array([-1]), np.array([-1])
print(f"mat_a[{indices0}, {indices1}] :\n {mat_a[indices0, indices1]}" )
```

```python
mat_a = np.arange(12).reshape(3, 4)
print(f"ndarray : \n{mat_a}")

indices0, indices1 = np.array([0, 1, 2]), np.array([1, 2, 3])

for idx0, idx1 in zip(indices0, indices1):
    print(f"({idx0}, {idx1})")

print(f"mat_a[{indices0}, {indices1}] :\n {mat_a[indices0, indices1]}" )


indices0 = np.array([[0, 1, 2], [0, 1, 2]])
indices1 = np.array([[0, 1, 2], [1, 2, 3]])

for row0, row1 in zip(indices0, indices1):
  for col0, col1 in zip(row0, row1):
      print(f"({col0}, {col1})", end= " ")
  print()  

print(f"mat_a[{indices0}, {indices1}] :\n {mat_a[indices0, indices1]}" )
```
