```python
# Rounding Sorting 
# numpy.sort() : value 
# numpy.argsort() : index 
# ascending 
# descending 

import numpy as np 

np.random.seed(30)
vec_x = np.random.randint(1, 100, (10, ))
print("===========Ascending===========")
print(f"vec_x : {vec_x}")

sort_x = np.sort(vec_x)
print(f"sort_x: {sort_x}")

arg_sort_x = np.argsort(vec_x)
print(f"arg_sort_x: {arg_sort_x}")

print("\n===========Descending===========")

vec_y = np.trunc(np.random.uniform(-10, 10, (10, )))
print(f"vec_y : {vec_y}")

sort_y = np.trunc(np.sort(vec_y)[::-1])
print(f"sort_y: {sort_y}")

arg_sort_y = np.argsort(vec_y)[::-1]
print(f"arg_sort_y: {arg_sort_y}")
```

```python
import numpy as np 

pred = np.random.uniform(0, 100, (5, ))
pred /= pred.sum()
print(f"pred {pred}")

top3_pred = np.sort(pred)[::-1][:3]
top3_indices = np.argsort(pred)[::-1][:3]
print(f"top3_pred {top3_pred}")
print(f"top3_indices {top3_indices}")
```

```python
np.random.seed(30)
mat_x = np.random.randint(0, 100, (4, 5))
print(f"mat_x \n{mat_x}")

print("===========Ascending===========")
sort_asc = np.sort(mat_x, axis=0)
arg_sort_asc = np.argsort(mat_x, axis=0)
# arg_sort_asc = np.argsort(mat_x, axis=1)

print(f"sort_asc \n{sort_asc}")
print(f"arg_sort_asc \n{arg_sort_asc}")

print("\n===========Descending===========")
sort_desc = np.sort(mat_x, axis=0)[::-1, :]
arg_sort_desc = np.argsort(mat_x, axis=0)[::-1, :]
# arg_sort_asc = np.argsort(mat_x, axis=1)

print(f"sort_desc \n{sort_desc}")
print(f"arg_sort_desc \n{arg_sort_desc}")

```

```python
import numpy as np 
np.random.seed(30)
score = np.random.randint(0, 100, (5, 3))
print(f"score : \n{score}")
sort_asc = np.sort(score, axis=0)
sort_desc = np.sort(score, axis=0)[::-1, :]
print(f"asc : \n{sort_asc}")
print(f"desc : \n{sort_desc}")

print("\n========Indicies========")
arg_sort_asc = np.argsort(score, axis=0)
arg_sort_desc = np.argsort(score, axis=0)[::-1, :]
print(f"asc : \n{arg_sort_asc}")
print(f"desc : \n{arg_sort_desc}")
```
