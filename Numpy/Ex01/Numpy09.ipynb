```python
import numpy as np 

vec_a = np.array([True, False, True, False])
print(f"vec_a shape : {vec_a.shape} ndim :  {vec_a.ndim} \n {vec_a}")

tup_nonzero = np.nonzero(vec_a)
tup_where = np.where(vec_a)
print(f"tup_nozero type: {type(tup_nonzero)} \n{tup_nonzero}")
print(f"tup_where type : {type(tup_where)} \n{tup_where}")
```

```python
import numpy as np
np.random.seed(30)
# Matrix 
mat_c = np.array([[1, 0, 7], 
                  [8, 1, 0], 
                  [1, 0, 0]])
print(f"mat_c shape : {mat_c.shape} ndim : {mat_c.ndim} : \n {mat_c}")
tup_nonzero = np.nonzero(mat_c)
trans_nonzero = np.transpose(np.nonzero(mat_c)) 
print(f"tup_nonzero type :{type(tup_nonzero)} : {tup_nonzero}")
print(f"trans_nonzero : {trans_nonzero}")

print("="*30)
# Tensor 
ten_a = np.array([[[True, False, True], 
                   [True, False, True]],
                  [[False, True, False],
                   [True, False, True]]])

print(f"ten_a shape : {ten_a.shape} ndim : {ten_a.ndim} : \n {ten_a}")

tup_nonzero = np.nonzero(ten_a)
trans_nonzero = np.transpose(np.nonzero(ten_a))

print(f"tup_nonzero type :{type(tup_nonzero)} : {tup_nonzero}")
print(f"trans_nonzero : {trans_nonzero}")
```

```python
import numpy as np 

np.random.seed(30)
mat_a = np.random.randint(-2, 3, size=(3, 3))
print(f"mat_a shape : {mat_a.shape} ndim : {mat_a.ndim}\n {mat_a}")

u_nonzero = mat_a[np.nonzero(mat_a)]
u_where = mat_a[np.where(mat_a)]
u_bool = mat_a[mat_a!=0]
print(f"u_nonzero shape : {u_nonzero.shape} ndim : {u_nonzero.ndim}\n {u_nonzero}")
print(f"u_where shape : {u_where.shape} ndim : {u_where.ndim}\n {u_where}")
print(f"u_bool shape : {u_bool.shape} ndim : {u_bool.ndim}\n {u_bool}")
```
