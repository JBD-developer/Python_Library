```python
# Making Coordinates 
# Repeating ndarrays 

# Coordinates from 1D vectors 

import numpy as np 

# y = x^2 -x Graph 
# z = x^2 +y^2 Graph 

x = np.arange(-2, 3)
y = np.arange(-2, 3)

print(f"x : {x}")
print(f"y : {y}")

X = x.reshape(1, -1).repeat(y.shape[0], axis=0)
print(f"X {X.shape} : \n{X}")
Y = y.reshape(-1, 1).repeat(x.shape[0], axis=1)
print(f"Y {Y.shape} : \n{Y}")

Z = np.square(X) + np.square(Y)
print(f"Z {Z.shape} : \n{Z}")
```

```python
import numpy as np 

x = np.arange(-2, 3)
y = np.arange(-2, 3)
print(f"x :{x.shape} \n{x}")
print(f"x :{y.shape} \n{y}")

X, Y = np.meshgrid(x, y)
print(f"X shape {X.shape} \n{X}")
print(f"Y shape {Y.shape} \n{Y}")

Z = np.square(X) + np.square(Y)
print(f"Z : \n{Z}")
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

print(f"x : {x.shape} \n{x}")
print(f"y : {y.shape} \n{y}")

X, Y = np.meshgrid(x, y)
Z = np.square(X) + np.square(Y)

fig= plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection="3d")

ax.plot_wireframe(X, Y, Z)
ax.tick_params(labelsize=20)
```
