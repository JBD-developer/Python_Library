```python
import numpy as np 

vec_x = np.arange(5)
vec_y = np.arange(2, 6)

print(f"vec_x : {vec_x}")
print(f"vec_y : {vec_y}")

for _x in vec_x : 
    print(_x + vec_y)
```

```python
import numpy as np 

vec_x = np.arange(5)
vec_y = np.arange(2, 6)

X, Y = np.meshgrid(vec_x, vec_y)

Z = X + Y 

X, Y, Z  = X.T, Y.T, Z.T
print(f"X \n{X}")
print(f"Y \n{Y}")
print(f"Z \n{Z}")
```

```python
import numpy as np 

x = np.arange(5)
y = np.arange(2, 6)
print(f"x : shape {x.shape} \n{x}")
print(f"y : shape {y.shape} \n{y}")

X = x.reshape(-1, 1)
print(f"X : shape {X.shape} \n{X}")
Y = y.reshape(1, -1)
print(f"Y shape {Y.shape} \n{Y}")

Z = X + Y 
print(f"Z shape {Z.shape} \n{Z}")
print(f"x[0] + y {Z[0, :]}")
print(f"x[3] + y {Z[3, :]}")
```

```python
# fully connected operations for vectors  
import numpy as np 
np.random.seed(30)
X = np.random.randint(-5, 5, (4, 2))
Y = np.random.randint(-5, 5, (3, 2))
print(f"X shape {X.shape} \n{X}")
print(f"Y shape {Y.shape} \n{Y}")
print("for loop")

# case for 
for x in X : 
    for y in Y : 
        add = x + y
        print(f"{add}", end="")     
    print()
```

```python
import numpy as np 

X = np.random.uniform(-5, 5, (4, 2))
Y = np.random.uniform(-5, 5, (3, 2))
print(f"X shape {X.shape}")
print(f"Y shape {Y.shape}")

print("=====expand_dims=====")

X = np.expand_dims(X, axis=1)
Y = np.expand_dims(Y, axis=0)
print(f"X shape {X.shape}")
print(f"Y shape {Y.shape}")
 
Z = X + Y
print(f"Z shape {Z.shape}") 
print(Z)

```
