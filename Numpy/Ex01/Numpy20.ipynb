```python
# Merging ndarrays using np.concatenate

import numpy as np 

vec_a = np.random.randint(0, 10, (3, ))
vec_b = np.random.randint(0, 10, (4, ))
print(f"vec_a shape : {vec_a.shape} \n {vec_a}")
print(f"vec_b shape : {vec_b.shape} \n {vec_b}")
# case two vector 
concat = np.concatenate((vec_a, vec_b))
print(f"concat : shape {concat.shape} \n{concat}") 

# case three vector 
vec_c = np.random.randint(0, 10, (4, ))
vec_d = np.random.randint(0, 10, (4, ))
vec_e = np.random.randint(0, 10, (4, ))
print(f"vec_a shape {vec_c.shape} {vec_c}")
print(f"vec_a shape {vec_d.shape} {vec_d}")
print(f"vec_a shape {vec_e.shape} {vec_e}")

concat = np.concatenate([vec_c, vec_d, vec_e], axis=0)
print(f"concat : shape {concat.shape} length {len(concat)} \n{concat}")
```

```python
import numpy as np 

dataset_tmp = list()

# vstack 
for iter in range(100):
    data_sample = np.random.uniform(0, 5, (1, 4))
    dataset_tmp.append(data_sample)

concat_axis0 = np.concatenate(dataset_tmp, axis=0)
print(f"concat_axis0 shape {concat_axis0.shape}")

# hstack 
dataset_tmp = list()
for iter in range(100):
    data_sample = np.random.uniform(0, 5, (4, 1))
    dataset_tmp.append(data_sample)

concat_axis1 = np.concatenate(dataset_tmp, axis=1)
print(f"concat_axis1 shape {concat_axis1.shape}")
    
```

```python
import numpy as np 
# Using dstack 
R = np.random.randint(0, 10, (100, 200))
G = np.random.randint(0, 10, size=R.shape)
B = np.random.randint(0, 10, size=R.shape)

image = np.dstack([R, G, B])

print(image.shape)
```
