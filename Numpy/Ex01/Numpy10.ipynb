```python
# Numpy axis and Keepdims Arguments 
import numpy as np 
# Vector case 1
"""

"""
print("\n============Vector============")
vec_a = np.arange(5)
print(f"vec_a shape : {vec_a.shape} ndim : {vec_a.ndim} \n {vec_a}")
print(f"vec_a.sum() : {vec_a.sum()}")
print(f"np.sum(vec_a) : {np.sum(vec_a)}")

# Matrix case 1
print("\n============Matrix============")
mat_a = np.arange(6).reshape(2, -1)
print(f"mat_a shape : {mat_a.shape} ndim : {mat_a.ndim} \n {mat_a}")
print(f"mat_a.sum() : {mat_a.sum()}")
print(f"np.sum(mat_a) : {np.sum(mat_a)}")

total = 0 
for row in range(len(mat_a)):
    for col in range(len(mat_a[row])):
        total += mat_a[row][col] 
print(f"total : {total}")

# Tensor case 1
print("\n============Matrix============")
ten_a = np.arange(12).reshape(2, 3, 2)
print(f"ten_a shape : {ten_a.shape} ndim : {ten_a.ndim} \n {ten_a}")
print(f"ten_a.sum() : {ten_a.sum()}")
print(f"np.sum(ten_a) : {np.sum(ten_a)}")
```

```python
import numpy as np 
mat_d = np.arange(12).reshape(3, 4)
print(f"mat_d shape : {mat_d.shape} ndim : {mat_d.ndim} \n{mat_d}")

sum_row_mat1 = mat_d.sum(axis=0, keepdims=True)
print("\n============Matrix | Keepdims=True ============")
print(f"sum_row_mat shape : {sum_row_mat1.shape} ndim : {sum_row_mat1.ndim} \n{sum_row_mat1}")

sum_row_mat2 = mat_d.sum(axis=0, keepdims=False)
print("\n============Matrix | Keepdims=False ============")
print(f"sum_row_mat shape : {sum_row_mat2.shape} ndim : {sum_row_mat2.ndim} \n{sum_row_mat2}")

sum_col_mat1 = mat_d.sum(axis=1, keepdims=True)
print("\n============Matrix | Keepdims=True ============")
print(f"sum_col_mat shape : {sum_col_mat1.shape} ndim : {sum_col_mat1.ndim} \n{sum_col_mat1}")

sum_col_mat2 = mat_d.sum(axis=1, keepdims=False)
print("\n============Matrix | Keepdims=False ============")
print(f"sum_col_mat shape : {sum_col_mat2.shape} ndim : {sum_col_mat2.ndim} \n{sum_col_mat2}")
```

```python
import numpy as np 

ten_a = np.arange(24).reshape(2 , 3, 4)

for dim in range(len(ten_a)):
    for row in range(len(ten_a[dim])):
        print()
        for col in range(len(ten_a[dim][row])):
            print(f"ten_a[{dim}][{row}][{col}] : {ten_a[dim][row][col]}", end= " ")
        print(end= " ")

sum_col = ten_a.sum(axis=2)
print("\n============Axis=2, Keepdims=False ============")
print(f"sum_col shapea{sum_col.shape} ndim {sum_col.ndim}\n {sum_col}")
sum_col2 = ten_a.sum(axis=2 , keepdims=True)
print("\n============Axis=2, Keepdims=True ============")
print(f"sum_col2 shapea{sum_col2.shape} ndim {sum_col2.ndim}\n {sum_col2}")
```

```python
import numpy as np 

np.random.seed(30)
num_test_time, num_student, num_subject = 4, 3, 4 
min_sccore, max_score =50, 100 
ten_score = np.random.randint(low = min_sccore, 
                              high = max_score,
                              size= (num_test_time,
                                     num_student, 
                                     num_subject))

print(f"ten_score shape : {ten_score.shape}, ndim : {ten_score.ndim} \n {ten_score}")
# subject student sum 
sum_score = ten_score.sum(axis=0)
# subject sum
sum_score = ten_score.sum(axis=1)
mean_score = ten_score.mean(axis=0)
print(f"mean_total :  shape {mean_score.shape} , ndim {mean_score.ndim} \n{mean_score}")
print(f"sum_total :  shape {sum_score.shape} , ndim {sum_score.ndim} \n{sum_score}")
```
