```python
import numpy as np 
# Indexing and Slicing ndarrays 
# Vector Indexing Slicing 
vec_a = np.array([1, 2, 3, 4, 5, 6, 7, 8])
print("ndarray : {} ".format(vec_a))
print("vec_a[0] : {} ".format(vec_a[0]))
print("vec_a[-1] : {} ".format(vec_a[-1]))    
print("vec_a[1:9]: {} ".format(vec_a[1:9]))   
print("vec_a[::-1]: {} ".format(vec_a[::-1]))

vec_b = np.arange(10)    
print("vec_b[0:3]: {} ".format(vec_b[0:3]))   
print("vec_b[3:7]: {} ".format(vec_b[3:7]))
print("vec_b[-2:]: {} ".format(vec_b[-2:]))
print("vec_b[:5] : {} ".format(vec_b[:5]))
print("vec_b[2:7:2]: {} ".format(vec_b[2:7:2]))
print("vec_b[::3]: {} ".format(vec_b[::3]))
print("vec_b[2:7:-1]: {} ".format(vec_b[7:2:-1]))
print("vec_b[8:3:-2]: {} ".format(vec_b[8:3:-2]))
print("vec_b[::-3]: {} ".format(vec_b[::-3]))

vec_c = np.arange(10)
indicies = np.array([0, 3, 6, -1])

print("vec_c[0, 3, 6, -1]: {} ".format(vec_c[[0, 3, 6, -1]]))
print("vec_c[indicies]: {} ".format(vec_c[indicies]))   
print("vec_c[vec_c % 3 == 0] : {}".format(vec_c[vec_c % 3 ==0]))

list_a = list(range(1, 11))
print(list_a)
for data_idex in range(5):
    list_a[data_idex] = 0

print(list_a)

np_a = np.arange(1, 11)
np_a[::2] =200
print(np_a)
np_a[5: -1:3] = 300
print(np_a)
```
