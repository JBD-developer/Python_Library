```python
# Memory Optimization 
# copy and view of ndarrays
# copy <-> View 
import copy 
import numpy as np 
original_arr = np.arange(5)
view_arr = original_arr.view()
print(original_arr)
print(view_arr)
original_arr += 10
print(original_arr)
print(view_arr)

a = np.arange(5)
b = a[0:3]
print(a)
print(b)
b[...] = 10
print(a)
print(b)

a = np.arange(5)
b = copy.deepcopy(a)
print(a)
print(b)

b[...] =10 

print(a)
print(b)

np_a = np.arange(5)
np_b = np_a.copy()
np_c = np_a.view()
np_d = np_a[0:3]

print(np_b.base is np_a)
print(np_c.base is np_a)
print(np_d.base is np_a)
```

```python
origianl_a = np.arange(4)
reshape_b = origianl_a.reshape((2, 2))
reshape_b[0, 0] =100

print(reshape_b.base is origianl_a ,"\n")
print(origianl_a)
print(reshape_b)
print("\n\n")

origianl_a = np.arange(6)
resize_b = np.resize(origianl_a, (4, 2))
resize_b[0, 1] =100

print(resize_b.base is origianl_a ,"\n")
print(origianl_a)
print(resize_b)
```
