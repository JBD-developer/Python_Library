```python
import numpy as np 

# Numpy Sum, Prod, Diff, and Statistics 

# numpy.cumsum()
# ndarray.cumsum()

vec_a = np.arange(5)
print(f"vec_a : \n {vec_a}")
cumsum = np.cumsum(vec_a)
print(f"np.cumsum(vec_a) : {np.cumsum(vec_a)}")
print(f"vec_a.cumsum() {vec_a.cumsum()}" )


mat_a = np.arange(9).reshape(3, 3)
print(f"mat_a shape {mat_a.shape}, ndim : {mat_a.ndim} \n{mat_a}")
mat_cumsum = mat_a.cumsum(axis=1)
mat_cumsum = mat_a.cumsum().reshape(3, 3)

print(f"mat_cumsum \n{mat_cumsum}")

```

```python
import numpy as np 

scores = np.random.normal(loc = 10, 
                          scale = 5, 
                          size = (100, ))

scores_var = np.var(scores)
scores_std = np.std(scores)

print(f"variance : {scores_var}")
print(f"standard deviation : {scores_std}")

print(f"square of std : {scores_std**2}")
print(f"square root of var : {scores_var**0.5}")
```

```python
import numpy as np

# Maxinum and Minimum
np.random.seed(30)

print("\n==============Vector==============")
vec_a = np.random.randint(1, 10, (10, ))
vec_b = np.random.randint(1, 10, (10, ))

print(f"vec_a shape {vec_a.shape} ndim : {vec_a.ndim} \n {vec_a}")
print(f"vec_b shape {vec_b.shape} ndim : {vec_b.ndim} \n {vec_b}")
maximum = np.maximum(vec_a, vec_b)
minimum = np.minimum(vec_a, vec_b)
print(f"maximum :\n {maximum}")
print(f"minimum :\n {minimum}")


print("\n==============Matrix==============")
mat_a = np.random.randint(1, 10, (3, 4))
mat_b = np.random.randint(1, 10, (3, 4))
print(f"mat_a shape {mat_a.shape} ndim : {mat_a.ndim} \n {mat_a}")
print(f"mat_b shape {mat_b.shape} ndim : {mat_b.ndim} \n {mat_b}")
maximum = np.maximum(mat_a, mat_b)
minimum = np.minimum(mat_a, mat_b)
print(f"maximum :\n {maximum}")
print(f"minimum :\n {minimum}")
```

```python
vec_a = np.random.randint(1, 10, (10, ))
vec_b = np.random.randint(1, 10, (10, ))

print(f"vec_a shape {vec_a.shape} ndim : {vec_a.ndim} \n {vec_a}")
print(f"vec_b shape {vec_b.shape} ndim : {vec_b.ndim} \n {vec_b}")

maximum = np.zeros_like(vec_a)
maximum[vec_a >= vec_b ] = vec_a[vec_a >= vec_b]
maximum[vec_a < vec_b ] = vec_b[vec_a < vec_b]

print(f"np.maxinum \n{np.maximum(vec_a, vec_b)}")
print(f"maximum \n {maximum}")
```

```python
np.random.seed(30)
u = np.random.randint(0, 10, (10, ))
v = np.random.randint(0, 10, (10, ))

print(f"u shpae: {u.shape} ndim : {u.ndim}")
print(f"v shpae: {v.shape} ndim : {v.ndim}")

up_val = np.full_like(u, fill_value=100)
down_val = np.full_like(v, fill_value=-100)
print(np.where(u > v, up_val, down_val))
```
