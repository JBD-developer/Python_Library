```python
# Overall Repetition 
# Repeating ndarrays 
# np.tile(A, reps)
import numpy as np 

vec_a = np.arange(4).reshape(-1, 1)
print(f"ndarray : shape{vec_a.shape} \n{vec_a}")

print("======repeat======")
rep_vec = np.repeat(vec_a, repeats=3, axis=1)
print(f"repeat ndarray  shape {rep_vec.shape} \n {rep_vec}")


print("======tile======")
vec_b = np.arange(4)
tile_vec = np.tile(vec_b, reps=3)
print(f"tile ndarray shape {tile_vec.shape} \n {tile_vec}")
```

```python
import numpy as np 

mat_a = np.arange(6).reshape(2, 3)
print(f"ndarray : shape{mat_a.shape} \n{mat_a}")

# row repeat once
# column repeat twice 
tile_mat = np.tile(mat_a, reps=[1, 2])
print(f"tile ndarray reps [1, 2]: shape{tile_mat.shape} \n{tile_mat}")

tile_mat = np.tile(mat_a, reps=[2, 1])
print(f"tile ndarray reps [2, 1]: shape{tile_mat.shape} \n{tile_mat}")

tile_mat = np.tile(mat_a, reps=[2, 2])
print(f"tile ndarray reps [2, 2]: shape{tile_mat.shape} \n{tile_mat}")
```
