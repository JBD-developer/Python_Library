```python
# Merging ndarrays 
# hstack vstack 
import numpy as np 

# vector 
vec_a = np.arange(8)
vec_b = np.arange(9, 17)
print(f"vec_a :{vec_a}")
print(f"vec_b :{vec_b}")

# vertical 
vstack = np.vstack([vec_a, vec_b])
print(f"vstack : shape {vstack.shape}\n{vstack}")

# horizontal 
hstack = np.hstack([vec_a, vec_b])
print(f"hstack : shape {hstack.shape} \n{hstack}")
```

```python
import numpy as np 
np.random.seed(30)
# row vector 
mat_a = np.random.randint(0, 10, (1, 3))
mat_b = np.random.randint(0, 10, (1, 3))

print(f"mat_a shape{mat_a.shape} :\n{mat_a}")
print(f"mat_b shape{mat_b.shape} :\n{mat_b}")

vstack = np.vstack((mat_a, mat_b))
print(f"vstack : shape {vstack.shape} \n{vstack}")


hstack = np.hstack((mat_a, mat_b))
print(f"hstack : shape {hstack.shape} \n{hstack}")
```

```python
import numpy as np 
np.random.seed(30)
# column vector 
mat_a =np.random.randint(0, 10, (3, 1))
mat_b =np.random.randint(0, 10, (3, 1))

print(f"mat_a shape{mat_a.shape} :\n{mat_a}")
print(f"mat_b shape{mat_b.shape} :\n{mat_b}")

vstack = np.vstack((mat_a, mat_b))
hstack = np.hstack((mat_a, mat_b))
print(f"vstack : shape {vstack.shape} \n{vstack}")
print(f"hstack : shape {hstack.shape} \n{hstack}")
```

```python
import numpy as np 
np.random.seed(30)
#matrix vector 
#vertical 
mat_a = np.random.randint(0, 10, (3, 4))
vec_a = np.random.randint(0, 10, (4, ))

print(f"mat_a shape{mat_a.shape} :\n{mat_a}")
print(f"vec_a shape{vec_a.shape} :\n{vec_a}")

vstack = np.vstack([vec_a, mat_a])
print(f"vstack : shape {vstack.shape} \n{vstack}")

```
