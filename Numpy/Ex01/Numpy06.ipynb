```python
mat_c = np.arange(12).reshape(4, 3)

for row in range(len(mat_c)):
    for col in range(len(mat_c[row])):
        print(f"mat_c[{row}][{col}]: {mat_c[row][col]}" ,end= "")
    print()
print("mat_c[1:, 0]:\n {}".format(mat_c[1:, 0])) 
print("mat_c[:3, 1]:\n {}".format(mat_c[:3, 1])) 
print("mat_c[1:3, 2]:\n {}".format(mat_c[1:3, 2]))       
```

```python
mat_d = np.arange(16).reshape(4, 4)

for row in range(len(mat_d)):
    for col in range(len(mat_d[row])):
        print(f"mat_d[{row}][{col}] : {mat_d[row][col]}", end="")
    print()
print()
print(f"mat_d[1:3, 1:3] :\n {mat_d[1:3, 1:3]}")
print(f"mat_d[:2, :3] :\n {mat_d[:2, :3]}")
print(f"mat_d[1:, -2:] :\n {mat_d[1:, -2:]}")
print(f"mat_d[-2:, :2] :\n {mat_d[-2:, :2]}")
print(f"mat_d[-2:, :-2] :\n {mat_d[-2:, :-2]}")
```

```python
import numpy as np 

mat_e = np.arange(16).reshape(4, 4)

for row in range(len(mat_e)):
    for col in range(len(mat_e[row])):
        print(f"[{row}][{col}] : {mat_e[row][col]}", end=" ")
    print()

print(f"mat_e[1:3, 1:3] :\n {mat_e[1:3, 1:3]}")
print(f"mat_e[:2, :3] :\n {mat_e[:2, :3]}")
print(f"mat_e[1:, -3:] :\n {mat_e[1:, -2:]}")
print(f"mat_e[-2:, :-2] :\n {mat_e[-2:, :-2]}")
```

```python
import numpy as np 

image = np.arange(9).reshape(3, 3)
print(f"ndarray: \n{image}\n")

horizontal_flip = image[:, ::-1]
print(f"horizotal_flip: \n{horizontal_flip}")

vertical_flip = image[::-1, :]
print(f"vertical_flip \n{vertical_flip}")

rotation = image[::-1, ::-1]
print(f"rotation \n{rotation}")
```
