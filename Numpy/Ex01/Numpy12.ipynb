```python
# Rounding and sorting 

import numpy as np 
# around
# round_
# round 

vec_n = np.random.uniform(-5, 5, (5, ))
print(f"vec_n : {vec_n}")

np_around = np.around(vec_n, decimals=2)
np_around_ = np.around(vec_n, decimals=2)
vec_round = vec_n.round(decimals=2)

print(f"np_around : {np_around}")
print(f"np_around_ : {np_around_}")
print(f"vec_round : {vec_round}")

```

```python
import numpy as np 

scores = np.random.uniform(0, 100, (100, 5))

means = scores.mean(axis=0)
stds = scores.std(axis = 0)

print(f"means : \n{means}")
print(f"means : \n{stds}")

print(f"means : \n{means.round(decimals=2)}")
print(f"stds : \n{stds.round(2)}")

```

```python
import numpy as np 

x = np.random.uniform(-5, 5, (5, ))

np_around = np.around(x)
np_round_ = np.round_(x)
x_round = x.round()

print(f"np_around : \n{np_around}")
print(f"np_round_ : \n{np_round_}")
print(f"x_round : \n{x_round}")
```

```python
import numpy as np 

# ceil 
# floor 
# truncation 
x = np.random.uniform(-5, 5, (5, ))
print(f"x : \n{x}")
x_ceil = np.ceil(x)
x_floor = np.floor(x)
x_trunc = np.trunc(x)

print(f"x_ceil : {x_ceil}")
print(f"x_floor : {x_floor}")
print(f"x_trunc : {x_trunc}")
```

```python
import numpy as np 

x = np.random.uniform(-5, 5, (5, ))
print(f"x : \n{x}")
trunc_where = np.where(x >=0, np.floor(x), np.ceil(x))
print(f"trunc_where : {trunc_where}")
```
