```python
import numpy as np 
# Element-wise Operations and Broadcasting 

np.random.seed(30)
a = np.random.randint(1, 5, (5, ))
b = np.random.randint(1, 5, (5, ))

print("a : ", a)
print("b : ", b)
print("a + b   : ", (a + b))
print("a - b   : ", (a - b))
print("a * b   : ", (a *  b))
print("a / b   : ", (a /  b))
print("a // b  : ", (a // b))
print("a % b   : ", (a %  b))
print("a ** b  : ", (a ** b))
print("\n\n")
print("a + b   : ", a.__add__(b))
print("a - b   : ", a.__sub__(b))
print("a * b   : ", a.__mul__(b))
print("a / b   : ", a.__truediv__(b))
print("a // b  : ", a.__floordiv__(b))
print("a % b   : ", a.__mod__(b))
print("a ** b  : ", a.__pow__(b))

print("a > b : " , a > b)
print("a > b : " , a >= b)
print("a > b : " , a < b)
print("a > b : " , a <= b)
print("a == b : " , a == b)
print("a != b : " , a != b)

mat_np1 = np.random.randint(1, 5, (2, 3))
mat_np2 = np.random.randint(1, 5, (2, 3))

print("mat_np1 :\n" , mat_np1)
print("mat_np2 :\n" , mat_np2)

print("mat_np1 + mat_np2 :\n", mat_np1 + mat_np2)
print("mat_np1 - mat_np2 :\n", mat_np1 - mat_np2) 

print("mat_np1 > mat_np2 :\n", mat_np1 > mat_np2)
print("mat_np1 >= mat_np2 :\n", mat_np1 >= mat_np2)

a = np.array(5)
mask = np.array([0, 1, 0, 1, 0])

print("input :", a)
print("mask :", mask)
print("output :", a*mask)
```

```python
# Broadcasting Cases 

# When ndims Are Equal(Matrices)

mat_np = np.arange(9).reshape(3, 3)
# vec_np = np.array([[0, 10, 20]])
# vec_np = 10 * np.arange(3).reshape(-1, 3)
vec_np = 10* np.arange(3).reshape(-1, 3)

print(mat_np)
print(vec_np)

print("mat_np{} / {}".format(mat_np.shape, mat_np.ndim))
print("vec_np{} / {}".format(vec_np.shape, vec_np.ndim))
print(mat_np + vec_np)

print("\n\n")
vec_np = 10 * np.arange(3).reshape(3, -1)
print(vec_np)


print("mat_np{} / {}".format(mat_np.shape, mat_np.ndim))
print("vec_np{} / {}".format(vec_np.shape, vec_np.ndim))
print(mat_np + vec_np)
```

```python
col_mat = np.arange(3).reshape(3, -1)
row_mat = np.arange(3).reshape(-1, 3)

print("col_mat {} / {}".format(col_mat.shape, col_mat.ndim))
print("row_mat {} / {}".format(row_mat.shape, row_mat.ndim))
print(col_mat)
print(row_mat)
```

```python
# Tensor Broadcasting 
ten_np1 = np.arange(18).reshape(2, 3, 3)
# print(ten_np1)
ten_np2 = 10*np.arange(9).reshape(1, 3, 3)
# print(ten_np2)
# print(ten_np1 + ten_np2)
# print("\n\n")

ten_np3 = 10*np.arange(6).reshape(2, 1, 3)
print(ten_np3)
print(ten_np1 + ten_np3)

ten_np4 = 10*np.arange(6).reshape(2, 3, 1)
print(ten_np4)
print(ten_np1 + ten_np4)
```

```python
# When ndims Are Not Equal 
# scalar vector 
a = np.array(3)
u = np.arange(1, 5)

print("a {}/{}".format(a.shape, a.ndim))
print("u {}/{}".format(u.shape, u.ndim))
print(a)
print(u)
print("a + u", a + u)
print("a - u", a - u)
print("a * u", a * u)
print("a // u", a // u)
print("a / u", a / u)
print("a % u", a & u)
print("a ** u", a ** u)

a = 10*np.array([1, 2])
b = np.arange(6).reshape(3, 2)
print(a)
print(b)
print(a + b)
```
