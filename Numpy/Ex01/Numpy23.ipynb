```python
# Repeating ndarrays 
# Application Repetition 
import numpy as np 

mat_x = np.arange(4*3).reshape(4, 3)
print(f"mat_x shape{mat_x.shape} \n{mat_x}")
sum_mat = mat_x.sum(axis=0, keepdims=True)
print(sum_mat)

rep_mat = sum_mat.repeat(repeats=3, axis=0)
print(rep_mat)

print(rep_mat.transpose())
```

```python
import numpy as np 

mat_x = np.arange(4*3).reshape(4, 3)
print(f"mat_x shape {mat_x.shape}\n{mat_x}")

mat_y = mat_x.sum(0, keepdims=True).repeat(3, 0).T
print(f"mat_y shape {mat_y.shape} \n{mat_y}")
```

```python
import numpy as np 
mat_x = np.arange(12).reshape(4, 3)
print(f"max_x shape {mat_x.shape} \n{mat_x}")

mat_x = mat_x.sum(1, keepdims=True).repeat(2, 1).T
print(mat_x)
```
