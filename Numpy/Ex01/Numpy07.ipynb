```python
import numpy as np 
# Indexing and Slicing ndarrays 
# Tensor Indexing Slicing 

ten_a = np.arange(27).reshape(3, 3, 3)
print(ten_a.ndim, ten_a.shape)

for dim in range(len(ten_a)):
    print()
    for row in range(len(ten_a[dim])):
        for col in range(len(ten_a[row])):
            print(f"ten_a[{dim}][{row}][{col}] : {ten_a[dim][row][col]}", end=" ")
        print() 
print()

# print(ten_a[0])
# print(ten_a[1])
# print(ten_a[2])

print("ten_a[0, 0] \n {}".format(ten_a[0, 0]))
print("ten_a[0, :2, :2] \n {}".format(ten_a[0, :2, :2]))
print("ten_a[1, :2, :] \n {}".format(ten_a[1, :2, :]))
print("ten_a[2, -2:, -2:] \n {}".format(ten_a[2, -2:, -2:]))
```

```python
ten_b = np.arange(3*3*3).reshape(3, 3, 3)
print(ten_b.ndim, ten_b.shape)

for dim in range(len(ten_b)):
    print()
    for row in range(len(ten_b[dim])):
        for col in range(len(ten_b[row])):
            print(f"ten_b[{dim}][{row}][{col}] : {ten_b[dim][row][col]}", end=" ")
        print() 
print()


print("ten_b[0, :2, :2] \n {}".format(ten_b[0, :2, :2]))
print("ten_b[1, :2, :2] \n {}".format(ten_b[1, :2, :]))
print("ten_b[2, -2:, -2:] \n {}".format(ten_b[2, -2:, -2:]))
```

```python
ten_c = np.arange(3*3*3).reshape(3, 3, 3)
print(ten_c.ndim, ten_c.shape)

for dim in range(len(ten_c)):
    print()
    for row in range(len(ten_c[dim])):
        for col in range(len(ten_c[row])):
            print(f"ten_c[{dim}][{row}][{col}] : {ten_c[dim][row][col]}", end=" ")
        print() 
print()

print("ten_c[:, :2, :] \n {}".format(ten_c[:, :2, :]))
print("ten_c[:, :, :1] \n {}".format(ten_c[:, :, :1]))
print("ten_c[1:, 1:, 1:] \n {}".format(ten_c[1:, 1:, 1:]))
```
