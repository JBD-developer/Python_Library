```python
import numpy as np 
score = [10, 20, 30, 40, 50]
# print("mean", np.mean(score))
# print("mean", np.var(score))
# a= np.array([1, 2, 3])

# print(type(a))
# for attr in dir(a):
#     print(attr)

u = [1, 2, 3]
v = [4, 5, 6]
w = [0] * len(u)

# python add 
for data_idx in range(len(u)):
    w[data_idx] = u[data_idx] + v[data_idx]
print(w)

# numpy add 
np_n = np.array(u)
np_v = np.array(v)
print(np_n + np_v)

M = [[1, 2, 3], 
     [4, 5, 6], 
     [7, 8, 9]]
N = [[11, 12, 13], 
     [14, 15, 16],
     [17, 18, 19]]
print(M + N)

N_row, N_col = len(M), len(M[0])

P = [[0] * N_col for _ in range(N_col)]


for row in range(N_row):
    for col in range(N_col):
        P[row][col] = M[row][col]+ N[row][col]

print(P)    
```
