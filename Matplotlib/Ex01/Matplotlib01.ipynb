#### Matplotlib
### 01. 기본 사용 
- plot() 하나의 숫자 리스트를 입력함으로 그래프를 그린다. 
- plot()함수는 line, marker 그래프 그리기에 사용되는 함수 
- plt.plot(parameter) 
- 기본적으로 plit([2, 3, 5, 10])로 입력하면 x 값은 [0, 1, 2, 3]이 된다. 
- 점 (0, 2), (1, 3), (2, 5), (3, 10)을 잇는 그래프를 표현한다. 
- show() 함수를 통해서 그래프를 화면에 나타낸다. 
- plt.show() 

### 02. 스타일 지정하기 
- plot() 함수를 통해 선의 형태와 색상을 지정하는 Format string을 세번째 인자에 입력 가능 
- axix() 함수를 통해 축의 범위[xmin, xmax, ymin, ymax]를 지정가능 하다.

```python
import matplotlib.pyplot as plt 
lst1 = [1, 2, 3, 4, 5, 6]
lst2 = [1, 4, 9, 16, 25, 36]
plt.plot(lst1)
plt.show()
plt.plot(lst1, lst2)
plt.show()
```

```python
lst1 = [1, 2, 3, 4]
lst2 = [1, 4, 9, 16]
plt.plot(lst1, lst2, 'y-')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

t = np.arange(0., 5., 0.2)
print(t)
plt.plot(t, t, 'r--', t, t** 2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
import matplotlib.pyplot as plt 

font1 = {"family":"consolas",
         "color": "b",
         "weight": "bold",
         "size":14}
font2 = {"family":"consolas",
         "color": "red",
         "weight": "normal",
         "size":14}

dic_data = {"data_x":[1, 2, 3, 4, 5], "data_y":[33, 4, 5, 6, 8]}
plt.plot('data_x', 'data_y', data=dic_data)
plt.xlabel("X-Axis", labelpad=20, fontdict=font1, loc='right')
# xlabel : left, center, right 
# ylabel : bottom, cnter, top
plt.ylabel("Y-Axis", labelpad=10, fontdict=font2, loc='center')
plt.title("Test")
plt.show()
```

### 03. 범례표시하기 
-범례(legend)는 그래프에 데이터의 종류를 표시하기 위한 텍스트이다. 
-legend()함수를 이용해서 그래프에 범례를 표시하는 것이 가능하다. 
-xlim(), ylim(), axis() 함수를 사용하면 그래프의 x축과 y축이 표시되는 범위를 지정할 수 있다. 

```python
import matplotlib.pyplot as plt 

dic_data = {"data_x":[1, 2, 3, 4, 5],
            "data_y":[22, 4, 5, 6, 7]}

plt.plot("data_x", "data_y", data=dic_data, label="Price ($)")
plt.legend(loc=(0.5, 0.5))
plt.xlabel("X-Axis", loc='center')
plt.ylabel("Y-Axis", loc="center")
plt.title("Test")
plt.show()
```
